```python
import cv2, cv
import serial
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ser=serial.Serial(port='/dev/ttyACM0',
                  baudrate=115200)
```

```python
def get_frame(ser):
    try:
        retu=np.array(ser.readline()[:-4].split(',')).astype(np.int8)/128.0
        if retu.size is not 128:
            return np.zeros((128))
        return retu
    except:
        return np.zeros((128))
```
